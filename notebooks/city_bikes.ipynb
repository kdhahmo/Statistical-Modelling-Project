using the [CityBikes API Documentation](https://api.citybik.es/v2/) how can I access information on 'free_bikes' and 'empty_slots'?

```python
# get city name and network id so that I can use it to make another request and have access to the information I need
CityNetworkRquest = requests.get('http://api.citybik.es/v2/networks?fields=location,id,name,href')
CityNetworkRquest
```

```python
# prepare to parse
CityNetworkRquestJSON = CityNetworkRquest.json()
# access name, city, endpoint, and id for next request. endpoint is to make the request, the other parts are for context
FlattenedCNR = [
    {'name': entry['name'],'city': entry['location']['city'],'endpoint': entry['href'],'id': entry['id']}
    for entry in CityNetworkRquestJSON['networks']
]
FlattenedCNR
```
